```python
df.plot(kind='scatter', x='Age', y='Tenure_Months', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)

#Kode di atas digunakan untuk membuat plot scatter yang menunjukkan hubungan antara usia ('Age') dan lama bekerja dalam bulan ('Tenure_Months') menggunakan pustaka Matplotlib dan pandas.
```

```python
plt.subplots(figsize=(8, 8))
df_2dhist = pd.DataFrame({
    x_label: grp['IsActiveMember'].value_counts()
    for x_label, grp in df.groupby('HasCrCard')
})
sns.heatmap(df_2dhist, cmap='viridis')
plt.xlabel('HasCrCard')
_ = plt.ylabel('IsActiveMember')


#visualisasi ini bertujuan untuk memberikan wawasan yang lebih dalam tentang hubungan antara kepemilikan kartu kredit dan keaktifan anggota, yang dapat digunakan untuk analisis lebih lanjut atau pengambilan keputusan.
```

```python
plt.figure(figsize=(10, 6))
for category in df['Credit_Score_Category'].unique():
    plt.scatter(df[df['Credit_Score_Category'] == category]['Balance'],
                df[df['Credit_Score_Category'] == category]['EstimatedSalary'],
                label=category, alpha=0.5)
plt.xlabel('Balance')
plt.ylabel('Estimated Salary')
plt.title('Customer Balance vs. Estimated Salary')
_ = plt.legend()


# isualisasi ini bertujuan untuk menunjukkan hubungan antara saldo dan gaji yang diperkirakan, serta bagaimana hubungan ini bervariasi antar kategori skor kredit.
```

```python
df.groupby('Gender').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)

# Visualisasi ini bertujuan untuk menampilkan distribusi jumlah individu berdasarkan jenis kelamin, sehingga memudahkan pemahaman tentang proporsi gender dalam dataset.
```

```python
df['CreditScore'].plot(kind='line', figsize=(8, 4), title='Credit Score')
plt.gca().spines[['top', 'right']].set_visible(False)

#Visualisasi ini bertujuan untuk menunjukkan tren skor kredit dari waktu ke waktu atau berdasarkan urutan data.
```

```python
df['Age'].plot(kind='line', figsize=(8, 4), title='Age')
plt.gca().spines[['top', 'right']].set_visible(False)


# Visualisasi ini bertujuan untuk menunjukkan tren usia dari waktu ke waktu atau berdasarkan urutan data.
```

```python
plt.figure(figsize=(20, 10))
sns.countplot(x='Age', data=df)
plt.title('Umur')
plt.xlabel('Umur')
plt.ylabel('Jumlah')
plt.show()

# Visualisasi ini bertujuan untuk menunjukkan sebaran jumlah individu dalam setiap kategori usia.
```

```python
sns.lineplot(x='Age', y='Balance', data=df, marker='o')
plt.title('Age VS Balance')
plt.xlabel('Age')
plt.ylabel('Balance')
plt.grid(True)
plt.show()

# Visualisasi ini bertujuan untuk menunjukkan hubungan antara usia dan saldo, apakah ada tren tertentu yang muncul.
```

```python
plt.figure(figsize=(8, 8))
df['Credit_Score_Category'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set3', explode=[0.05]*df['Geography'].nunique())
plt.title('Tipe Credit')
plt.ylabel('')  # Menghilangkan label sumbu Y untuk pie chart
plt.show()


# Visualisasi ini bertujuan untuk menunjukkan proporsi masing-masing kategori skor kredit dalam total data.
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Loyalty_Score'], bins=20, kde=True)
plt.title('Loyalty')
plt.xlabel('Loyalty Score')
plt.ylabel('Frekuensi')
plt.show()

#Visualisasi ini bertujuan untuk menunjukkan sebaran nilai skor loyalitas dalam data.
```

```python
plt.figure(figsize=(10, 10))
sns.scatterplot(x='Tenure_Months', y='Balance', data=df, hue='IsActiveMember')
plt.title('Balance vs Tenure Month Berdasarkan Isactivemember')
plt.xlabel('TMI')
plt.ylabel('Balance')
plt.show()

# Visualisasi ini bertujuan untuk menunjukkan hubungan antara lama berlangganan dan saldo, serta bagaimana keanggotaan aktif mempengaruhi hubungan tersebut.
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Credit_Score_Category', y='Balance', data=df)
plt.title('Balance Berdasarkan Tipe Credit')
plt.xlabel('Tipe Kredit')
plt.ylabel('Balance')
plt.show()

# digunakan untuk membuat boxplot yang menunjukkan distribusi saldo berdasarkan kategori skor kredit.
```

```python
# Memilih hanya kolom numerik untuk korelasi
numerical_df = df.select_dtypes(include=['float64', 'int64'])

# Visualisasi Korelasi Antar Variabel Numerik
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_df.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Matriks Korelasi Variabel Numerik')
plt.show()

# untuk menunjukkan tingkat korelasi antara variabel numerik dalam dataset.
```
